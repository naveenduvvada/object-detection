```python
import matplotlib.pyplot as plt
```

```python
config_file = 'C:/Users/duvva/Downloads/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model ='C:/Users/duvva/Downloads/frozen_inference_graph.pb'
```

```python
print(frozen_model)
```

```python
import os
if not os.path.exists(frozen_model):
    print(f"File not found: {frozen_model}")
else:
    print(f"File found: {frozen_model}")
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels = []
file_name = 'C:/Users/duvva/Downloads/coco.names'
with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
#read an image

img = cv2.imread('C:/Users/duvva/Downloads/wp6693787.jpg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex,confidece,bbox = model.detect(img,confThreshold = 0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img, boxes, (255, 0, 0), 2)
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
     raise IOError("cannot open webcame")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret,frame = cap.read()
  
    ClassIndex,confidece,bbox = model.detect(frame,confThreshold=0.55)
    
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd,conf,boxes in zip(ClassIndex.flatten(),confidece.flatten(),bbox):
            if (ClassInd<=80):
                cv2.rectangle(frame,boxes,(225,0,0),2)
                cv2.putText(frame,classLabels[ClassInd-1],(boxes[0]+1,boxes[1]+2),font,fontScale=font_scale,color=(0,225,0),thickness=3)

    cv2.imshow('object Detection Tutorial',frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
         break

cap.release()
cv2.destroyAllWindows()
```
